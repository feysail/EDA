```python
cols=['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'Tamb', 'RH', 'WS',
       'WSgust', 'WSstdev', 'WD', 'WDstdev', 'BP', 'Cleaning', 'Precipitation',
       'TModA', 'TModB']
df[cols]=df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df.describe().columns
```

```python
fig,axes=plt.subplots(5,4,figsize=(30,25))
axes=axes.flatten()
for i, feature in enumerate(df.describe().columns):
    if feature=='Timestamp':
        continue
    else:
         df.groupby(df['Timestamp'].dt.strftime('%B'))[feature].sum().sort_values().plot(kind='line',ax=axes[i])
         plt.ylabel(feature)
    
```

```python
fig, axes=plt.subplots(5, 4, figsize=(30,25))
axes=axes.flatten()
list=['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'Tamb', 'RH', 'WS',
       'WSgust', 'WSstdev', 'WD', 'WDstdev', 'BP', 'Cleaning', 'Precipitation',
       'TModA', 'TModB', 'Comments']
for i,column in enumerate(list):
    df.groupby(df['Timestamp'].dt.strftime('%B'))[column].sum().sort_values().plot(kind='line',ax=axes[i])
    
```

```python
plt.figure(figsize=(10,10))
df.groupby('Cleaning')[['ModA','ModB']].sum().plot(kind='line')
plt.show()
```

we can infer from the above data that  Cleaning has a significant effect on sensors
